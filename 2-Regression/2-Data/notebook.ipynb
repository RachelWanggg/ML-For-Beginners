```python
import pandas as pd
import  matplotlib.pyplot as plt
```

```python
pumpkins = pd.read_csv('../data/US-pumpkins.csv')
pumpkins = pumpkins[pumpkins['Package'].str.contains('bushel', case=True, regex=True)]
pumpkins.head()
```

```python
pumpkins.isnull().sum()
```

```python
columns_to_select = ['City Name', 'Low Price', 'High Price', 'Package']
pumpkins = pumpkins.loc[:, columns_to_select]
```

```python
price = (pumpkins['Low Price'] + pumpkins['High Price']) / 2

pumpkins = pd.DataFrame({'Origin': pumpkins['City Name'], 'Package': pumpkins['Package'], 'Low Price': pumpkins['Low Price'],'High Price': pumpkins['High Price'], 'Price': price})


pumpkins.loc[pumpkins['Package'].str.contains('1 1/9'), 'Price'] = price/(1 + 1/9)

pumpkins.loc[pumpkins['Package'].str.contains('1/2'), 'Price'] = price/(1/2)

print(pumpkins)
```

```python
pumpkins.groupby(['Origin'])['Price'].mean().plot(kind='bar')
plt.ylabel("Pumpkin Price")
```
